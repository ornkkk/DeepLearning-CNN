```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import MaxPooling2D, Dense, Conv2D, Flatten, BatchNormalization, GlobalAvgPool2D
from tensorflow.keras.layers import PReLU, ReLU, Dropout, Input, Concatenate, AveragePooling2D, Add
from tensorflow.keras import Sequential
from sklearn.model_selection import KFold
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.initializers import Constant
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import img_to_array, array_to_img
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.applications.resnet50 import preprocess_input as preprocess_input_resnet

from tensorflow.keras import optimizers
from tensorflow.keras import callbacks
```

```python
from numba import cuda 
device = cuda.get_current_device()
device.reset()
```

```python
(train_X, train_y), (test_X, test_y) = fashion_mnist.load_data()

validity_set = False

if validity_set:
    val_X, train_X = train_X[0:6000], train_X[6000:]
    val_y, train_y = train_y[0:6000], train_y[6000:]

# summarize loaded dataset
print("Train: X=%s, y=%s" % (train_X.shape, train_y.shape))
print("Test: X=%s, y=%s" % (test_X.shape, test_y.shape))

# plot first few images
for i in range(9):
    # define subplot
    plt.subplot(330 + 1 + i)
    # plot raw pixel data
    plt.imshow(train_X[i], cmap=plt.get_cmap("gray"))
    
# show the figure
plt.show()
```

### Useful Functons anf callbacks

```python
reduce_learning = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=2,
    verbose=1,
    mode='auto',
    min_delta=0.0001,
    cooldown=2,
    min_lr=0)

eary_stopping = callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=7,
    verbose=1,
    mode='auto',
    restore_best_weights=True)

def checkpoint(file_name="model.h5"):
    checkpoint_cb = callbacks.ModelCheckpoint(file_name, save_best_only=True)
    return checkpoint_cb

channel = 1 # set to 1 or 3

def evaluate_model(model, X_train, y_train, epoch=10):
    history = model.fit(X_train, y_train, epochs=epoch,
                        validation_split=0.1, callbacks=[reduce_learning, eary_stopping, checkpoint(model.name+".h5")], verbose=1)
    
    return history

def plot_diagnostics_asis(history):
    pd.DataFrame(history.history).plot(figsize=(8, 5))
    plt.grid(True)
    plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
    plt.show()
    
def plot_diagnostics(history):
    '''
    the validation error is computed at the end of
    each epoch, while the training error is computed using a running mean during each
    epoch. So the training curve should be shifted by half an epoch to the left.
    '''
    
    epoch = len(history.history["accuracy"])
    epochs=np.arange(0, epoch)

    plt.title('History')
    plt.plot(epochs-0.5, history.history['loss'], color='blue', label='train_loss')
    plt.plot(epochs, history.history['val_loss'], color='green', label='val_loss')
    plt.plot(epochs-0.5, history.history['accuracy'], color='orange', label='train_accuracy')
    plt.plot(epochs, history.history['val_accuracy'], color='red', label='val_accuracy')
    plt.grid(True)
    plt.xlim([0, epoch-0.5])
    plt.ylim([0, 1.5])
    plt.legend()
    plt.show()
```

### Preparing Data

```python
def prepare_data_MLP(train_X, test_X):
    train_X = train_X / 255.
    test_X = test_X / 255.
    
    return train_X, test_X

def prepare_data_CNN(train_X, test_X, train_y, test_y, channels=1):
    train_X = np.dstack([train_X.reshape([train_X.shape[0], 784])] * channels)
    test_X = np.dstack([test_X.reshape([test_X.shape[0], 784])] * channels)

    train_X = train_X.reshape(-1, 28,28,channels)
    test_X= test_X.reshape (-1,28,28,channels)
    
    train_X = train_X.astype('float32')
    test_X = test_X.astype('float32')
    
    #normalize to range 0-1
    train_X = train_X / 255.
    test_X = test_X / 255.
    
    train_y = to_categorical(train_y)
    test_y = to_categorical(test_y)
    
    return train_X, test_X, train_y, test_y

def prepare_data_VGG(train_X, test_X, train_y, test_y, channels=1):
    train_X = np.dstack([train_X.reshape([train_X.shape[0], 784])] * channels)
    test_X = np.dstack([test_X.reshape([test_X.shape[0], 784])] * channels)

    train_X = train_X.reshape(-1, 28,28,channels)
    test_X= test_X.reshape (-1,28,28,channels)
    
    train_X = np.asarray([img_to_array(array_to_img(im, scale=False).resize((48,48))) for im in train_X])
    test_X = np.asarray([img_to_array(array_to_img(im, scale=False).resize((48,48))) for im in test_X])
    
    if channels==3:
        train_X = preprocess_input(train_X)
        test_X  = preprocess_input(test_X)

    train_y = to_categorical(train_y)
    test_y = to_categorical(test_y)
    
    train_X = train_X.astype('float32')
    test_X = test_X.astype('float32')
    
    #normalize to range 0-1
    train_X = train_X / 255.
    test_X = test_X / 255.
    
    return train_X, test_X, train_y, test_y
    
def prepare_data_ResNet(train_X, test_X, train_y, test_y, channels=1):
    train_X = np.dstack([train_X.reshape([train_X.shape[0], 784])] * channels)
    test_X = np.dstack([test_X.reshape([test_X.shape[0], 784])] * channels)

    train_X = train_X.reshape(-1, 28,28,channels)
    test_X= test_X.reshape (-1,28,28,channels)
    
    if channels==3:
        train_X = preprocess_input_resnet(train_X)
        test_X  = preprocess_input_resnet(test_X)

    train_y = to_categorical(train_y)
    test_y = to_categorical(test_y)
    
    train_X = train_X.astype('float32')
    test_X = test_X.astype('float32')
    
    #normalize to range 0-1
    train_X = train_X / 255.
    test_X = test_X / 255.
    
    return train_X, test_X, train_y, test_y
    
    
```

### Baseline MLP Model

```python
train_X_mlp, test_X_mlp = prepare_data_MLP(train_X, test_X)
train_y_mlp, test_y_mlp = train_y, test_y
```

```python
history_mlp = evaluate_model(model_mlp, train_X_mlp, train_y_mlp, epoch=100)
```

```python
plot_diagnostics(history_mlp)
```

```python
model_mlp.evaluate(test_X_mlp, test_y_mlp)
```

```python
def cnn_model_1():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', padding="same", use_bias=True, kernel_initializer='he_uniform',
                     input_shape=(28, 28, channel)))
    model.add(MaxPooling2D((2, 2)))
    #model.add(Dropout(rate=0.45))
    model.add(BatchNormalization())
    model.add(Conv2D(32, (3, 3), activation='relu', padding="same", use_bias=True, kernel_initializer='he_uniform'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
    ##model.add(Dropout(rate=0.25))
    
    model.add(Conv2D(64, (3, 3), activation='relu', padding="same", use_bias=True, kernel_initializer='he_uniform',
                    kernel_regularizer="l2"))
    model.add(MaxPooling2D((2, 2)))
    #model.add(Dropout(rate=0.45))
    model.add(BatchNormalization())
    model.add(Conv2D(64, (3, 3), activation='relu', padding="same", use_bias=True, kernel_initializer='he_uniform',
                    kernel_regularizer="l2"))
    model.add(BatchNormalization())
    model.add(MaxPooling2D((2, 2)))
    ##model.add(Dropout(rate=0.25))
    
    model.add(Flatten())
    model.add(Dense(128, activation='relu', kernel_initializer='he_uniform', kernel_regularizer="l2", use_bias=True))
    model.add(Dropout(rate=0.25))
    model.add(Dense(10, activation='softmax', kernel_initializer='he_uniform', use_bias=True))
    
    # compile model
    #opt = SGD(lr=0.01, momentum=0.9)
    #opt = SGD()
    model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

### VGG-16

```python
def VGG_Block(in_layer, n_filters, n_conv, prelu_activation=False, batch_normalization=False, dropout=False):
    
    for _ in range(n_conv):
        in_layer = Conv2D(n_filters, (3,3), padding="same", kernel_initializer="he_normal", use_bias=True)(in_layer)
        in_layer = PReLU()(in_layer) if prelu_activation else ReLU()(in_layer)
        in_layer = BatchNormalization()(in_layer) if batch_normalization else in_layer
        
    in_layer = MaxPooling2D((2,2), strides=(2,2))(in_layer)
    in_layer = Dropout(rate=0.40)(in_layer) if dropout else in_layer
    
    return in_layer


def Create_VGG16(input_layer, prelu_activation=False, batch_normalization=False, dropout=False, classes=10):
    
    n_filters = (16, 32, 64, 256, 256) #(64, 128, 256, 512, 512)
    n_conv = (2, 2, 3, 3, 3)
    layer = input_layer
    
    for block in zip(n_filters, n_conv):
        layer = VGG_Block(layer, block[0], block[1], prelu_activation, batch_normalization, dropout)
    
    
    layer = Flatten()(layer)
    
    for _ in range(2):
        layer = Dense(1024, kernel_initializer="he_normal", use_bias=True,
                    kernel_regularizer="l2")(layer)
        layer = PReLU()(layer) if prelu_activation else ReLU()(layer)
        layer = BatchNormalization()(layer) if batch_normalization else layer
        layer = Dropout(rate=0.25)(layer) if dropout else layer
    
    layer = Dense(classes, kernel_initializer="he_normal", activation="softmax", use_bias=True,
                    kernel_regularizer="l2")(layer)
    
    model = Model(inputs=input_layer, outputs=layer)
    return model
```

```python
train_X_vgg, test_X_vgg, train_y_vgg, test_y_vgg = prepare_data_VGG(train_X, test_X, train_y, test_y, channels=channel)

train_X_vgg.shape, test_X_vgg.shape, train_y_vgg.shape, test_y_vgg.shape
```

```python
input_layer = Input(shape=(48, 48, channel))
model_vgg = Create_VGG16(input_layer, prelu_activation=True, batch_normalization=True, dropout=True)
model_vgg._name ="VGG-16"

#model.summary()

#opt = SGD()
model_vgg.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])

history_vgg = evaluate_model(model_vgg, train_X_vgg, train_y_vgg, epoch=100)
```

### ResNet-50

```python
def Residual_Module(in_layer, K, stride=(1,1), bottleneck=True, pre_activation=True, red=False):
    
    layer = in_layer
    if bottleneck:
        layer = BatchNormalization()(layer) if pre_activation else layer
        layer = ReLU()(layer) if pre_activation else layer
        layer = Conv2D(K//4, (1,1), strides=stride, kernel_initializer="he_normal",
                       bias_initializer=Constant(value=0.2), padding="same", activation="linear")(layer)
        layer = BatchNormalization()(layer) if not pre_activation else layer
        layer = ReLU()(layer) if not pre_activation else layer
        
        layer = BatchNormalization()(layer) if pre_activation else layer
        layer = ReLU()(layer) if pre_activation else layer
        layer = Conv2D(K//4, (3,3), kernel_initializer="he_normal",
                       bias_initializer=Constant(value=0.2), padding="same", activation="linear")(layer)
        layer = BatchNormalization()(layer) if not pre_activation else layer
        layer = ReLU()(layer) if not pre_activation else layer
        
        layer = BatchNormalization()(layer) if pre_activation else layer
        layer = ReLU()(layer) if pre_activation else layer
        layer = Conv2D(K, (1,1), kernel_initializer="he_normal",
                       bias_initializer=Constant(value=0.2), padding="same", activation="linear")(layer)
        layer = BatchNormalization()(layer) if not pre_activation else layer
        layer = ReLU()(layer) if not pre_activation else layer
        
    else:
        layer = BatchNormalization()(layer) if pre_activation else layer
        layer = ReLU()(layer) if pre_activation else layer
        layer = Conv2D(K, (3,3), strides=stride, kernel_initializer="he_normal",
                       bias_initializer=Constant(value=0.2), padding="same", activation="linear")(layer)
        layer = BatchNormalization()(layer) if not pre_activation else layer
        layer = ReLU()(layer) if not pre_activation else layer
        
        layer = BatchNormalization()(layer) if pre_activation else layer
        layer = ReLU()(layer) if pre_activation else layer
        layer = Conv2D(K, (3,3), kernel_initializer="he_normal",
                       bias_initializer=Constant(value=0.2), padding="same", activation="linear")(layer)
        layer = BatchNormalization()(layer) if not pre_activation else layer
        layer = ReLU()(layer) if not pre_activation else layer
        

    shortcut = Conv2D(K, (1,1), strides=stride, kernel_initializer="he_normal",
                    bias_initializer=Constant(value=0.2), padding="same", activation="linear")(in_layer) if red else in_layer
    
    out_layer = Add()([layer, shortcut])
    out_layer = ReLU()(out_layer)
    
    return out_layer
        

def Create_ResNet50(input_layer):
    stages = (3, 4, 6, 3)
    filters = (256, 512, 1024, 2048)
    
    layer = Conv2D(64, (7,7), strides=(2,2), kernel_initializer="he_normal",
                   bias_initializer=Constant(value=0.2), padding="same", activation="linear")(input_layer)
    layer = BatchNormalization()(layer)
    layer = ReLU()(layer)
    
    layer = MaxPooling2D((3,3), strides=(2,2))(layer)
    
    for i in range(len(stages)):
        stride = (1,1) if i==0 else (2, 2)
        layer = Residual_Module(layer, filters[i], stride, bottleneck=True, pre_activation=False, red=True)
        
        for j in range(stages[i]-1):
            layer = Residual_Module(layer, filters[i], pre_activation=False)
    
    layer = GlobalAvgPool2D()(layer)
    layer = Dense(10, activation="softmax")(layer)
    
    model = Model(inputs=input_layer, outputs=layer)
    return model
```

```python
train_X_res, test_X_res, train_y_res, test_y_res = prepare_data_ResNet(train_X, test_X, train_y, test_y, channels=channel)

train_X_res.shape, test_X_res.shape, train_y_res.shape, test_y_res.shape
```

```python
input_layer = Input(shape=(28, 28, channel))
model_res = Create_ResNet50(input_layer)
model_res._name ="ResNet-50"

#model.summary()

#opt = SGD()
model_res.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])

history_res = evaluate_model(model_res, train_X_res, train_y_res, epoch=100)
```

```python
plot_diagnostics(history_res)
```

```python
model_res.evaluate(test_X_res, test_y_res)
```
