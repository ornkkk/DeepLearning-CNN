```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, PReLU, ReLU, Conv3D, MaxPooling3D, Dropout, BatchNormalization, Input, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.utils import plot_model
import pydot, graphviz
```

```python
def VGG_Block(in_layer, n_filters, n_conv, prelu_activation=False, batch_normalization=False, dropout=False):
    
    for _ in range(n_conv):
        in_layer = Conv2D(n_filters, (3,3), padding="same", kernel_initializer="he_normal")(in_layer)
        in_layer = PReLU()(in_layer) if prelu_activation else ReLU()(in_layer)
        in_layer = BatchNormalization()(in_layer) if batch_normalization else in_layer
        
    in_layer = MaxPooling2D((2,2), strides=(2,2))(in_layer)
    in_layer = Dropout(rate=0.25)(in_layer) if dropout else in_layer
    
    return in_layer


def Create_VGG16(input_layer, prelu_activation=False, batch_normalization=False, dropout=False):
    
    n_filters = (64, 128, 256, 512, 512)
    n_conv = (2, 2, 3, 3, 3)
    layer = input_layer
    
    for block in zip(n_filters, n_conv):
        layer = VGG_Block(layer, block[0], block[1], prelu_activation, batch_normalization, dropout)
    
    #layer = VGG_Block(input_layer, 64, 2, prelu_activation, batch_normalization, dropout)
    #layer = VGG_Block(layer, 128, 2, prelu_activation, batch_normalization, dropout)
    #layer = VGG_Block(layer, 256, 3, prelu_activation, batch_normalization, dropout)
    #layer = VGG_Block(layer, 512, 3, prelu_activation, batch_normalization, dropout)
    #layer = VGG_Block(layer, 512, 3, prelu_activation, batch_normalization, dropout)
    
    layer = Flatten()(layer)
    
    for _ in range(2):
        layer = Dense(4096, kernel_initializer="he_normal")(layer)
        layer = PReLU()(layer) if prelu_activation else ReLU()(layer)
        layer = BatchNormalization()(layer) if batch_normalization else layer
        layer = Dropout(rate=0.25)(layer) if dropout else layer
    
    #layer = Dense(4096, kernel_initializer="he_normal")(layer)
    #layer = PReLU()(layer) if prelu_activation else ReLU()(layer)
    #layer = BatchNormalization()(layer) if batch_normalization else layer
    #layer = Dropout(rate=0.25)(layer) if dropout else layer
    
    layer = Dense(1000, kernel_initializer="he_normal", activation="softmax")(layer)
    
    model = Model(inputs=input_layer, outputs=layer)
    return model

def Create_VGG19(input_layer, prelu_activation=False, batch_normalization=False, dropout=False):
    
    n_filters = (64, 128, 256, 512, 512)
    n_conv = (2, 2, 4, 4, 4)
    layer = input_layer
    
    for block in zip(n_filters, n_conv):
        layer = VGG_Block(layer, block[0], block[1], prelu_activation, batch_normalization, dropout)
    
    layer = Flatten()(layer)
    
    for _ in range(2):
        layer = Dense(4096, kernel_initializer="he_normal")(layer)
        layer = PReLU()(layer) if prelu_activation else ReLU()(layer)
        layer = BatchNormalization()(layer) if batch_normalization else layer
        layer = Dropout(rate=0.25)(layer) if dropout else layer
    
    layer = Dense(1000, kernel_initializer="he_normal", activation="softmax")(layer)
    
    model = Model(inputs=input_layer, outputs=layer)
    return model
```

```python
input_layer = Input(shape=(224, 224, 3))
model = Create_VGG19(input_layer, prelu_activation=True, batch_normalization=True, dropout=True)
model.summary()
```

```python
plot_model(model, show_shapes=True, to_file='VGG-19.png')
```
