```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, PReLU, ReLU, Dropout, BatchNormalization
from tensorflow.keras.layers import Input, Dense, Flatten, Concatenate, AveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.utils import plot_model
from tensorflow.keras.initializers import Constant
import pydot, graphviz
```

```python
def Inception_Module(in_layer, num1x1, num3x3Reduce, num3x3, num5x5Reduce, num5x5, num1x1Pool):
    
    layer_1 = Conv2D(num1x1, (1,1), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(in_layer)
    
    layer_2 = Conv2D(num3x3Reduce, (1,1), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(in_layer)
    layer_2 = Conv2D(num3x3, (3,3), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(layer_2)
    
    layer_3 = Conv2D(num5x5Reduce, (1,1), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(in_layer)
    layer_3 = Conv2D(num5x5, (5,5), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(layer_3)

    layer_4 = MaxPooling2D((3,3), strides=(1,1), padding="same")(in_layer)
    layer_4 = Conv2D(num1x1Pool, (1,1), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(layer_4)
    
    out_layer = Concatenate()([layer_1, layer_2, layer_3, layer_4])
    
    return out_layer

def Create_GoogLeNet(input_layer):
    
    layer = Conv2D(64, (7,7), strides=(2,2), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(input_layer)
    layer = MaxPooling2D((3,3), strides=(2,2), padding="same")(layer)
    layer = tf.keras.layers.Lambda(tf.nn.local_response_normalization)(layer)
    
    layer = Conv2D(64, (1,1), strides=(1,1), padding="valid", activation="relu",
                     bias_initializer=Constant(value=0.2))(layer)
    layer = Conv2D(192, (3,3), strides=(1,1), padding="same", activation="relu",
                     bias_initializer=Constant(value=0.2))(layer)
    layer = tf.keras.layers.Lambda(tf.nn.local_response_normalization)(layer)
    layer = MaxPooling2D((3,3), strides=(2,2), padding="same")(layer)
    
    layer = Inception_Module(layer, 64, 96, 128, 16, 32, 32)
    layer = Inception_Module(layer, 128, 128, 192, 32, 96, 64)
    
    layer = MaxPooling2D((3,3), strides=(2,2), padding="same")(layer)
    
    layer = Inception_Module(layer, 192, 96, 208, 16, 48, 64)
    layer = Inception_Module(layer, 160, 112, 224, 24, 64, 64)
    layer = Inception_Module(layer, 128, 128, 256, 24, 64, 64)
    layer = Inception_Module(layer, 112, 144, 288, 32, 64, 64)
    layer = Inception_Module(layer, 256, 160, 320, 32, 128, 128)
    
    layer = MaxPooling2D((3,3), strides=(2,2), padding="same")(layer)
    
    layer = Inception_Module(layer, 256, 160, 320, 32, 128, 128)
    layer = Inception_Module(layer, 384, 192, 384, 48, 128, 128)
    
    layer = AveragePooling2D((7,7), strides=(1,1), padding="valid")(layer)
    layer = Dropout(rate=0.40)(layer)
    layer = Dense(1000, activation="softmax")(layer)
    
    model = Model(inputs=input_layer, outputs=layer)
    return model
```

```python
input_layer = Input(shape=(224, 224, 3))
model = Create_GoogLeNet(input_layer)
model.summary()
```

```python
plot_model(model, show_shapes=True, to_file='GoogLeNet.png')
```
